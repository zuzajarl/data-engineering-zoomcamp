```python
start = pd.Timestamp("2025-11-01")
end   = pd.Timestamp("2025-12-01")

trips_nov = df_taxi[(df_taxi['lpep_pickup_datetime'] >= start) & (df_taxi['lpep_pickup_datetime']  < end)]
trips_nov
```

```python
short_trips = trips_nov[trips_nov['trip_distance'] <= 1.0]

#First value is the number of rows matching the task 3 requirements 
short_trips.shape
```

```python
trips_filtered = df_taxi[df_taxi['trip_distance'] < 100.0]
trips_filtered
```

```python
longest_trip_date = trips_filtered[trips_filtered['trip_distance'] == trips_filtered.trip_distance.max()]['lpep_pickup_datetime']
longest_trip_date
```

```python
df_taxi = df_taxi.merge(
    df_zones[['LocationID', 'Zone']],
    left_on='PULocationID',
    right_on='LocationID',
    how='left'
)
df_taxi = df_taxi.drop(columns=['LocationID'])
```

```python
df_taxi
```

```python
start = pd.Timestamp("2025-11-18")
end   = pd.Timestamp("2025-11-19")
df_18th_nov = df_taxi[(df_taxi['lpep_pickup_datetime'] >= start) & (df_taxi['lpep_pickup_datetime'] < end)]
df_18th_nov
```

```python
trips_per_zone = df_18th_nov.groupby('Zone')['lpep_pickup_datetime'].count()
trips_per_zone = trips_per_zone.rename("trip_count").reset_index()

max_row_df = trips_per_zone.nlargest(1, 'trip_count')
print(max_row_df)
```

```python
start = pd.Timestamp("2025-11-01")
end   = pd.Timestamp("2025-12-01")

trips_nov = df_taxi[(df_taxi['lpep_pickup_datetime'] >= start) & (df_taxi['lpep_pickup_datetime']  < end)]

trips_nov = trips_nov.rename(columns={'Zone': 'PUZone'}) 

trips_nov = trips_nov.merge(
    df_zones[['LocationID', 'Zone']],
    left_on='DOLocationID',
    right_on='LocationID',
    how='left'
).rename(columns={'Zone': 'DOZone'}) 

trips_nov = trips_nov.drop(columns=['LocationID'])

ehn_df = trips_nov[trips_nov['PUZone'] == 'East Harlem North']
ehn_df
```

```python
tips_per_zone = ehn_df.groupby('DOZone')['tip_amount'].sum().rename("tip_sum").reset_index()
max_tip_per_zone = tips_per_zone[tips_per_zone['tip_sum'] == tips_per_zone.tip_sum.max()]
max_tip_per_zone
```
